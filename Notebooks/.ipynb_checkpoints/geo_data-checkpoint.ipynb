# Import

```python
import pandas as pd
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option('chained_assignment',None)
```

```python
fpath_listings_detailed = '../SanDiego/listings_detailed.csv'
list_df = pd.read_csv(fpath_listings_detailed)
print list_df.shape
```

```python
# 
```

```python
location_info = ['neighbourhood_cleansed','latitude','longitude']

features = ['price'] + location_info
print features

df = list_df[features]
df.head()
```

```python
def price_to_float(x):
    x = x.replace('$','')
    x = x.replace(',','')
    return float(x)

df['price'] = df['price'].apply(price_to_float)

print df.shape
df.head()
```

# Filter

```python
df_500 = df[df['price']>=500]
df_500.head()
```

# Plot

```python
fig,ax = plt.subplots(figsize=(8,6))
x = list(df['latitude'])
y = list(df['longitude'])
price = list(df['price'])
ax.scatter(x,y,c=price)
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(8,6))
lats = list(df_500['latitude'])
lons = list(df_500['longitude'])
price = list(df_500['price'])
ax.scatter(lats, lons, c=price)
plt.show()
```

```python
import gmaps
import gmplot
```

```python
gmap = gmplot.GoogleMapPlotter(32.75, -117.2, 11)
gmap.heatmap(lats, lons)
gmap.draw("heatmap_san_diego_airbnb.html")
```
