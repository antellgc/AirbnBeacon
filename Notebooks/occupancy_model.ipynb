# Import requirements

```python
import pandas as pd
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option('chained_assignment',None)
```

# Import calendar data

```python
def quick_csv_read(fpath):
    '''
    quickly read a file into a pandas dataframe, display its number of rows
    and columns, show the header
    '''
    df = pd.read_csv(fpath)
    print df.shape
    return df

fpath_calendar = '../SanDiego/calendar.csv'
calendar_df = quick_csv_read(fpath_calendar)
calendar_df.head()
```

# Survey calendar info

```python
new_dict = {'listing_id':[], 'number_available':[], 'number_unavailable':[]}

list_ids = calendar_df.listing_id
for lid in list(set(list_ids)):
    sub_df = calendar_df[calendar_df['listing_id']==lid]
    available_count = sum(sub_df['available']=='t')
    not_available_count = sum(sub_df['available']=='f')
    
    new_dict['listing_id'].append(lid)
    new_dict['number_available'].append(available_count)
    new_dict['number_unavailable'].append(not_available_count)
    
pd.DataFrame(new_dict)
```
